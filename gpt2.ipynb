# Draw Histogram of Weight Distribution (ALL LAYERS)

```python
import matplotlib.pyplot as plt
```

```python
# Extract the weight tensors from the state_dict
weights = []
for name, param in model.named_parameters():
    if 'weight' in name:
        weights.append(param.view(-1))
```

```python
# Concatenate the weight tensors into a single 1D array
flattened_weights = torch.cat(weights)
```

```python
import numpy as np
```

```python
# Plot a histogram of the weight distribution
x_values = np.arange(len(flattened_weights))
plt.hist(flattened_weights.detach().numpy(), bins=10000)
plt.xlabel('Weight Value')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
# Set the x-axis limits for the zoomed-in view
xmax = 1
xmin = -(xmax)
plt.xlim(xmin, xmax)  # Specify the desired minimum and maximum x-axis values
plt.show()
```
